```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import preprocessing
from sklearn.cross_validation import KFold, cross_val_score

dataset = pd.read_csv('Numeric.csv')
'''
print(dataset)
'''

#encode categorical variable
'''
print("The transformed Type is:",transfomed_Q9)
'''
#predict Type given Q1 ( )
CV =  dataset.Type.reshape((len(dataset.Type), 1))

data = (dataset.ix[:,'Q1':'Q12'].values).reshape((len(dataset.Type), 12))
'''
data = dataset.Q9.reshape((len(dataset.Q9), 1))
print(data)
print(CV)
'''
# Create linear regression object
#
DT = DecisionTreeClassifier(criterion="entropy", max_depth=3,min_samples_leaf = 2)

# Train the model using the training sets
DT.fit(data, CV)

# the model
with open("predict_Type.dot", 'w') as f:
    f = tree.export_graphviz(DT, out_file=f, feature_names=["Q1","Q2","Q3","Q4","Q5","Q6","Q7","Q8","Q9","Q10","Q11","Q12"], class_names=["Luxury apartment","Shared Apartments","House","Dorms","Single bedroom apartment"], filled=True)



#predict the class for each data point
predicted = DT.predict(data)
print("Predictions: \n", np.array([predicted]).T)

# predict the probability/likelihood of the prediction
print("Probability of prediction: \n",DT.predict_proba(data))

print("Feature importance: ", DT.feature_importances_)

print("Accuracy score for the model: \n", DT.score(data,CV))

print(metrics.confusion_matrix(CV, predicted, labels=["Luxury apartment","Shared Apartments","House","Dorms","Single bedroom apartment"]))

# Calculating 5 fold cross validation results
model = DecisionTreeClassifier()
kf = KFold(len(CV), n_folds=2)
scores = cross_val_score(model, data, CV, cv=kf)
print("MSE of every fold in 5 fold cross validation: ", abs(scores))
print("Mean of the 5 fold cross-validation: %0.2f" % abs(scores.mean()))

print("Which type is suggested, if resondant chooses to do Netflix and chill in his free time and prefers to stay at corner of cafe: ", DT.predict([1,0]),DT.predict_proba([1,0]))
```
