```python
import pandas as pd
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
import numpy as np
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn import preprocessing


dataset = pd.read_csv('book.csv')
print(dataset)
#pre-processing features
Q9 =  dataset.Q9.reshape((len(dataset.Q9), 1))

#encode categorical variable
enc = preprocessing.OneHotEncoder()
enc.fit(Q9)
transfomed_Q9 = enc.transform(Q9).toarray()
print("The transformed attend class is:",transfomed_Q9)

#predict Type given Q1
CV =  dataset.Type.reshape((len(dataset.Q9), 1))
data = dataset.Q9.reshape((len(dataset.Q9), 1))
NB = GaussianNB()

# Train the model using the training sets
NB.fit(data, CV)

#Model
print("Probability of the classes: ", NB.class_prior_)
print("Mean of each feature per class:\n", NB.theta_)
print("Variance of each feature per class:\n", NB.sigma_)
#predict the class for each data point
predicted = NB.predict(data)
print("Predictions:\n",np.array([predicted]).T)

# predict the probability/likelihood of the prediction
prob_of_pred = NB.predict_proba(data)
print("Probability of each class for the prediction: \n",prob_of_pred)

print("Accuracy of the model: ",NB.score(data,CV))

print("The confusion matrix:\n", metrics.confusion_matrix(CV, predicted, ['Luxury apartment','Shared Apartments','House','Dorms','Single bedroom apartment']))

# Calculating 5 fold cross validation results
model = GaussianNB()
kf = KFold(len(CV), n_folds=5)
scores = cross_val_score(model, data, CV, cv=kf)
print("MSE of every fold in 5 fold cross validation: ", abs(scores))
print("Mean of the 5 fold cross-validation: %0.2f" % abs(scores.mean()))

print("Which type is suggested, if resondant chooses to do Netflix and chill in his free time: ", NB.predict(2),NB.predict_proba(2))

```
